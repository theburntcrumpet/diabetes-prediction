# Train Diabetes Model

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("diabetes_risk_prediction_dataset.csv")
data
```

```python
yes_no_columns = [
    "Polyuria",
    "Polydipsia",
    "sudden weight loss",
    "weakness",
    "Polyphagia",
    "Genital thrush",
    "visual blurring",
    "Itching",
    "Irritability",
    "delayed healing",
    "partial paresis",
    "muscle stiffness",
    "Alopecia",
    "Obesity"
]

data["Gender"] = data["Gender"].replace({"Male": 0, "Female": 1})
data[yes_no_columns] = data[yes_no_columns].replace({"No": 0, "Yes": 1})
data["class"] = data["class"].replace({"Negative": 0, "Positive": 1})
data
```

```python
data_features = data.copy()
data_labels = data_features.pop("class")
```

```python
diabetes_model = keras.Sequential([
    keras.layers.Normalization(),
    keras.layers.Dense(16),
    keras.layers.Dense(128),
    keras.layers.Dense(1, activation="sigmoid")
])

diabetes_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

diabetes_model.fit(data_features,data_labels,epochs=16)

diabetes_model.evaluate(data_features, data_labels)
```

```python
result = diabetes_model.predict(np.array([40,0,0,1,0,1,0,0,0,1,0,1,0,1,1,1]))


diabetes_model.save("diabetes.keras")
result[0][0]
```
